```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
shark_data = pd.read_csv('SharkAttack.csv')
```

```python
# Cleaning the data and counting shark attacks by country
shark_attacks_by_country = shark_data['Country'].value_counts()
```

```python
# Extracting top 10 countries with most shark attacks
top_10_countries = shark_attacks_by_country.head(10)
```

```python
# Creating Line Chart
plt.figure(figsize=(10, 6))
top_10_countries.plot(kind='line', marker='o', color='blue')
plt.title('Top 10 Countries with Shark Attacks')
plt.xlabel('Country')
plt.ylabel('Number of Shark Attacks')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Creating Pie Chart
plt.figure(figsize=(8, 8))
top_10_countries.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Shark Attacks by Top 10 Countries')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Creating Bar Chart
plt.figure(figsize=(10, 6))
top_10_countries.plot(kind='bar', color='green')
plt.title('Top 10 Countries with Shark Attacks')
plt.xlabel('Country')
plt.ylabel('Number of Shark Attacks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
